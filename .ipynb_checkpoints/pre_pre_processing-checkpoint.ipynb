```python
import pandas as pd
import numpy as np
```

```python
def read_excel(data):
    df = pd.read_excel('Data Prep.xlsx', index_col=0, sheet_name=data)
    return df

def delete_x_row(df,x):
    df = df[x:]
    return df

```

```python
x_up = read_excel('x_up')
x_up = delete_x_row(x_up,1)
x_up
```
