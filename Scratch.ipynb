```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Data Prep.xlsx', index_col=0, sheet_name='Biaya_Jarak_Teus')
```

```python
col_list = list(df.columns)
```

```python
index_list = list(df.index)
```

```python
all_list = col_list+index_list
```

```python
all_port = list(set(all_list))
```

```python
new_df = pd.DataFrame(columns = all_port,
                     index = all_port)
new_df
```

```python
for col in col_list:
    for idx in index_list:
        data = df[col][idx]
        if pd.isna(data) == False:
#             print(data)
            new_df[col][idx] = data

new_df
            
```

```python
new_df.to_csv('Biaya_Jarak_Teus.csv')
```
